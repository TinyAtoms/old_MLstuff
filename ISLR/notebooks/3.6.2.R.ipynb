```r
library(MASS)
```

`symmary(model)` reports basic statistics

```r
# create linear model and fit with response~predictors
lm.fit = lm(medv~lstat, data=Boston)
summary(lm.fit)

names(lm.fit)
coef(lm.fit)
```

```r
# confidence interval
confint(lm.fit)
```

```r
# confidence interval for lstat[5,10,15]
predict(lm.fit, data.frame(lstat=c(5, 10, 15)), interval="confidence")
```

```r
# prediction interval for lstat[5,10,15]
predict(lm.fit, data.frame(lstat=c(5, 10, 15)), interval="predict")
```

```r
# plot y and X
plot(medv ~ lstat, data = Boston, pch=20)
# plot regression line
abline(lm.fit,  lwd =3 , col ="blue")
```

# Multiple regression
the function is `lm(y∼x1+x2+x3)`

```r
model = lm(medv~lstat+age, data=Boston)
summary(model)
```

# or just `lm(y~., data=DF)` to include all other rows

```r
model = lm(medv~., data=Boston)
summary(model)
plot(model)
```

We can use the vif function in the car package to check for correlation of features

```r
library(car)
vif(model)
```

```r
model = lm(medv~.-tax, data=Boston)
```

```r
plot(model)
```

# As we see, there's a clear nonlinearity. Especially when the graph is drawn

quick note:
y~x1:x2 does y ~ x1 * x2  
  
  
y~x1 * x2 does y ~ x1,x2, `x1 * x2`

```r
model2 = lm(medv~.+lstat:age, data=Boston)
summary(model2)
plot(model2)
```

```r
model3 = lm(medv ~ lstat + I(lstat^2), data=Boston)
## or this
model4 = lm(medv~lstat + poly(lstat,2), data=Boston)
```

```r
anova(model, model2, model3, model4)
```
