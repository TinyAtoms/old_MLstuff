```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
college = pd.read_csv("./../datasets/College.csv")
college.set_index("Unnamed: 0", inplace=True)
college.columns
```

```python
college.describe()
```

```python
college.head()
```

```python
ax = sns.violinplot(x="Private", y="Outstate", data=college)
```

```python
ax = sns.boxplot(x="Private", y="Outstate", data=college)
```

```python
sep =pd.cut(college.Top10perc, pd.interval_range(start=0, end=100, periods=2), labels=["Not elite", "Elite"])
college["Elite"] = sep 
```

```python
ax = sns.violinplot(x="Elite", y="Outstate", data=college)
```
